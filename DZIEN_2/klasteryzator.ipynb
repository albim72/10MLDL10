```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import kmeans_plusplus
```

```python
n_samples = 15_000
n_components = 5
```

```python
X,y_true = make_blobs(n_samples=n_samples,centers=n_components,cluster_std=1.2,random_state=1)
X = X[:,::-1]
```

```python
centers_init,indices = kmeans_plusplus(X,n_clusters=5,random_state=1)
```

```python
plt.figure()
colors = ["#4EACC5","#FF9C34","#4A9A06","m","c","#AA6575","#1111EE"]
for k,col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data,0],X[cluster_data,1],c=col,marker=".",s=10)
plt.scatter(centers_init[:,0],centers_init[:,1],c="b",s=50)
plt.title("algorytm k-means")
plt.xticks([])
plt.yticks([])
plt.show()
```
